# Abstract 
Goal is to predict whether a stock price in the S&P500 will move up or down based on a tweet. To do this a dataset composed of financial tweets will be combined with a dataset of the stock prices of 10 popular stocks from the S&P. This will be accomplished with a classification model such as KNN. Performance will be measured by the percentage of cases that the model can correctly predict from the test dataset.



# Background

This question is relevant because with companies like Robin-Hood and M1-Finance popping up in recent times and making stock market investing accessible to anyone with the knowledge to download a smartphone app, that opens the doors to social media marketing tactics that could potentially move prices by creating a mob of retail investors that act in unision. This is enhanced with the introduction of fractional investing offered through these platforms which allows an investor to purchase a fractional amount of a share, removing the barrier for low-capital-retail-investors to invest in companies that have a high price per share. These tactics can be any combination of covert/overt and unintentional/intentional to the human eye but perhaps there are underlying patterns that the right machine learning algorithm can caputre.

# Problem Statement

Classify/predict the next day effect of a tweet on the stock price of certain top stocks in the S&P500, to determine whether it will go up or down during the next trading day.

# Discussion

### Interpreting the result

The results of this investigation can be interpreted as there being a correlation between the masses on twitter and the next day price trend. The accuracy of the model decreases as the value of K increases which tells me that there is not an abundance of valuable data to help the model make its prediction rather there are a couple pieces of high quality data that can account for the majority of the decision. This downward trend was present in the models of every stock I looked at meaning that there is a high amount of generalizability for this conclusion.

The results of the confusion matrix's were less consistent with the balance of false positives to false negatives jumping back and forth from stock to stock. For that reason no generalizations can be made in this area, however it can be said that a model that leans more heavily towards false negatives is more useful to the average trader because it would lead to less losses than a model that leans towards false positives.

### Limitations

It would be fruitful to explore more models, due to the nature of this investigation it was hard to get to more than one due to the lengthy cleaning process. More data is also of great use, as it stands this investigation spans a 3-4 month period of time in 2020, which is a lot of data to begin with (almost a million tweets). However certain macro cycles in the stock markets like bull/bear can last longer than a 3-4 month period and influence the overall trend of most stocks, leading to more correct choices in a bull market by consistently predicting price increase. This is a likely explanation for the declining accuracy rate when increasing K.

### Ethics & Privacy
